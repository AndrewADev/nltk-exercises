### Exercise 5)

Humor has the higher lexical diversity score, at just over 23% compared to just over 12% for romance fiction.

Formulating these numbers another way, each (unique) word appears slightly more than 4 times (100 / 23.1) in the studied texts of the humor genre, while words in the studied texts in romantic fiction appear slighty more than 8 times each (100 / 12.1).
